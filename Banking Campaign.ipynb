### Insights from Correlation Matrix

```python
#Let's analyze the Pearson correlation matrix to extract the best insights from it:

#Age and Balance (Correlation: 0.1123):
#There is a weak positive correlation between age and balance.
#Insight: Older customers tend to have slightly higher balances, which is a common trend but not a strong relationship.

#Day and Campaign (Correlation: 0.137):
#There is a weak positive correlation between the day of the month and the number of contacts during the campaign.
#Insight: Certain days of the month might see more campaign activity, possibly due to strategic planning or customer availability.

#Campaign and Pdays (Correlation: -0.1027):
#There is a weak negative correlation between the number of contacts during the campaign and the number of days since the customer was last contacted.
#Insight: Customers contacted more frequently during the campaign might have been contacted more recently in previous campaigns.

#Pdays and Previous (Correlation: 0.5073):
#There is a moderate positive correlation between the number of days since the customer was last contacted and the number of contacts performed before this campaign.
#Insight: Customers who were contacted more in previous campaigns are more likely to have been contacted recently in this campaign.

#Previous and Pdays (Correlation: 0.5073):
#Same correlation value as above, but it indicates a bidirectional relationship.
#Insight: Customers with a higher number of previous contacts are also more likely to have been contacted recently in this campaign.
```

### Feature engineering

```python
# Create age groups
data['age_group'] = pd.cut(data['age'], bins=[0, 30, 40, 50, 60, 100], labels=['<30', '30-39', '40-49', '50-59', '60+'])

# Create balance categories
data['balance_category'] = pd.qcut(data['balance'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Create duration categories
data['duration_category'] = pd.qcut(data['duration'], q=4, labels=['Short', 'Medium', 'Long', 'Very Long'])

# Display the updated dataset with new features
print("\nUpdated dataset with new features:")
data.head()
```

### Based on the information gathered from the univariate analysis of different variables, we can derive insights regarding who keeps deposits in the bank or not. Let's analyze how it might relate to deposit:

Job:
1.Management, technician, admin, and services jobs seem to have higher frequencies, indicating that these occupations are more prevalent among the customers.
2.We can't directly infer deposit behavior from job titles alone, but it's worth noting that certain professions may have higher income levels or stability, potentially influencing deposit behavior.

Marital Status:
1.Married individuals have the highest frequency, followed by singles and divorced.
2.Marital status might have an impact on financial decisions, with married individuals possibly more inclined towards joint financial planning and deposits.

Education:
1.Customers with secondary and tertiary education levels are more frequent in the dataset.
2.Higher education levels might be associated with better financial literacy and planning, potentially leading to more deposits.

Housing Loan:
1.A significant number of customers do not have a housing loan.
2.Having fewer financial obligations like loans might make customers more likely to keep deposits in the bank.

Personal Loan:
1.The majority of customers do not have a personal loan.
2.Similar to housing loans, fewer financial commitments may contribute to deposit behavior.

Contact Method:
1.Most contacts were made via cellular communication.
2.The preferred contact method might indicate customer demographics and preferences, potentially influencing deposit behavior.

Month of Contact:
1.May, August, and July are the top months for contact, while December has the lowest frequency.
2.Seasonality or campaign timing might influence deposit subscriptions during specific months.

Previous Outcome:
1.A large number of previous outcomes are unknown, followed by failures and successes.
2.Past campaign outcomes may play a role in customer receptiveness to new offers like deposits.
